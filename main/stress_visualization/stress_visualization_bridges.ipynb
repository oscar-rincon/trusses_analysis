## Bridges

```python
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
warren_bridge_disp = analysis(warren_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
warren_bridge_esfuerzos = calc_esfuerzos_int(
    warren_bridge_data["nodes"], 
    warren_bridge_data["elements"], 
    warren_bridge_data["mats"], 
    warren_bridge_disp
)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
pratt_bridge_disp = analysis(pratt_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
pratt_bridge_esfuerzos = calc_esfuerzos_int(
    pratt_bridge_data["nodes"], 
    pratt_bridge_data["elements"], 
    pratt_bridge_data["mats"], 
    pratt_bridge_disp
)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
howe_bridge_disp = analysis(howe_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
howe_bridge_esfuerzos = calc_esfuerzos_int(
    howe_bridge_data["nodes"], 
    howe_bridge_data["elements"], 
    howe_bridge_data["mats"], 
    howe_bridge_disp
)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
double_warren_bridge_disp = analysis(double_warren_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
double_warren_bridge_esfuerzos = calc_esfuerzos_int(
    double_warren_bridge_data["nodes"], 
    double_warren_bridge_data["elements"], 
    double_warren_bridge_data["mats"], 
    double_warren_bridge_disp
)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
x_truss_bridge_disp = analysis(x_truss_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
x_truss_bridge_esfuerzos = calc_esfuerzos_int(
    x_truss_bridge_data["nodes"], 
    x_truss_bridge_data["elements"], 
    x_truss_bridge_data["mats"], 
    x_truss_bridge_disp
)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
k_truss_bridge_disp = analysis(k_truss_bridge_data, verbose=False)
# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
k_truss_bridge_esfuerzos = calc_esfuerzos_int(
    k_truss_bridge_data["nodes"], 
    k_truss_bridge_data["elements"], 
    k_truss_bridge_data["mats"], 
    k_truss_bridge_disp
)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(8, 5))

# Visualizar los esfuerzos en la cercha del puente Warren
axs[0, 0].set_title("Warren Truss Bridge", fontsize=8)
plt.sca(axs[0, 0])
vis_esfuerzos(
    warren_bridge_data["nodes"], 
    warren_bridge_data["elements"], 
    warren_bridge_esfuerzos, 
    desp=None
)

# Visualizar los esfuerzos en la cercha del puente Pratt
axs[0, 1].set_title("Pratt Truss Bridge", fontsize=8)
plt.sca(axs[0, 1])
vis_esfuerzos(
    pratt_bridge_data["nodes"], 
    pratt_bridge_data["elements"], 
    pratt_bridge_esfuerzos, 
    desp=None
)

# Visualizar los esfuerzos en la cercha del puente Howe
axs[1, 0].set_title("Howe Truss Bridge", fontsize=8)
plt.sca(axs[1, 0])
vis_esfuerzos(
    howe_bridge_data["nodes"], 
    howe_bridge_data["elements"], 
    howe_bridge_esfuerzos, 
    desp=None
)

axs[1, 1].set_title("Duble Warren Truss Bridge", fontsize=8)
plt.sca(axs[1, 1])
vis_esfuerzos(
    double_warren_bridge_data["nodes"], 
    double_warren_bridge_data["elements"], 
    double_warren_bridge_esfuerzos, 
    desp=None
)

# Visualizar los esfuerzos en la cercha del puente K
axs[2, 0].set_title("X Truss Bridge", fontsize=8)
plt.sca(axs[2, 0])
vis_esfuerzos(
    x_truss_bridge_data["nodes"], 
    x_truss_bridge_data["elements"], 
    x_truss_bridge_esfuerzos, 
    desp=None
)

# Visualizar los esfuerzos en la cercha del puente K
axs[2, 1].set_title("K Truss Bridge", fontsize=8)
plt.sca(axs[2, 1])
vis_esfuerzos(
    k_truss_bridge_data["nodes"], 
    k_truss_bridge_data["elements"], 
    k_truss_bridge_esfuerzos, 
    desp=None
)



plt.tight_layout()
plt.savefig("figs/stress_bridges.png", dpi=300)
plt.show()
```
