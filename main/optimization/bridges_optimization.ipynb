```python
import numpy as np
import matplotlib.pyplot as plt   
from datos_cerchas import warren_bridge_data, pratt_bridge_data, howe_bridge_data, double_warren_bridge_data, x_truss_bridge_data, k_truss_bridge_data
from plane_trusses import analysis
from analisis_cerchas import calc_esfuerzos_int, optimizar_cercha, vis_esfuerzos_secciones
```

A continuación se usa la función `minimize` de `scipy` para optimizar las secciones de los elementos de las cerchas. Esta función permite encontrar los valores óptimos de las variables que minimizan una función objetivo, sujeta a restricciones. En este caso, la función objetivo es la que calcula la masa total de la cercha, y las restricciones incluyen los límites de esfuerzo y las dimensiones mínimas y máximas de las secciones transversales.

El proceso de optimización se realiza para diferentes tipos de puentes, incluyendo el puente Warren, el puente Pratt, el puente Howe, el puente Warren doble y el puente tipo K. Para cada tipo de puente, se ajustan las secciones de los elementos para minimizar la masa total mientras se aseguran de que los esfuerzos en los elementos no excedan los límites permitidos y que las secciones transversales se mantengan dentro de los rangos especificados.

A continuación se presenta el código utilizado para optimizar y analizar cada tipo de puente.

```python
"""""
 Optimizar y analizar el puente Warren
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
warren_bridge_data_optimized, warren_bridge_masa_minima = optimizar_cercha(warren_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)

# Realizar el análisis estructural para obtener los desplazamientos de los nodos
warren_bridge_disp = analysis(warren_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
warren_bridge_esfuerzos = calc_esfuerzos_int(
    warren_bridge_data_optimized["nodes"], 
    warren_bridge_data_optimized["elements"], 
    warren_bridge_data_optimized["mats"], 
    warren_bridge_disp
)

"""""
 Optimizar y analizar el puente Pratt
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
pratt_bridge_data_optimized, pratt_bridge_masa_minima = optimizar_cercha(pratt_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)
# Realizar el análisis estructural para obtener los desplazamientos de los nodos
pratt_bridge_disp = analysis(pratt_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
pratt_bridge_esfuerzos = calc_esfuerzos_int(
    pratt_bridge_data_optimized["nodes"], 
    pratt_bridge_data_optimized["elements"], 
    pratt_bridge_data_optimized["mats"], 
    pratt_bridge_disp
)

"""""
 Optimizar y analizar el puente Howe
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
howe_bridge_data_optimized, howe_bridge_masa_minima = optimizar_cercha(howe_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)                                                                       

# Realizar el análisis estructural para obtener los desplazamientos de los nodos
howe_bridge_disp = analysis(howe_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
howe_bridge_esfuerzos = calc_esfuerzos_int(
    howe_bridge_data_optimized["nodes"], 
    howe_bridge_data_optimized["elements"], 
    howe_bridge_data_optimized["mats"], 
    howe_bridge_disp
)

"""""
 Optimizar y analizar el puente Warren doble
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
double_warren_bridge_data_optimized, double_warren_bridge_masa_minima = optimizar_cercha(double_warren_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)                                                                                         

# Realizar el análisis estructural para obtener los desplazamientos de los nodos
double_warren_bridge_disp = analysis(double_warren_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
double_warren_bridge_esfuerzos = calc_esfuerzos_int(
    double_warren_bridge_data_optimized["nodes"], 
    double_warren_bridge_data_optimized["elements"], 
    double_warren_bridge_data_optimized["mats"], 
    double_warren_bridge_disp
)

"""""
 Optimizar y analizar el puente tipo K
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
x_truss_bridge_data_optimized, x_truss_bridge_masa_minima = optimizar_cercha(x_truss_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)                                                                             

# Realizar el análisis estructural para obtener los desplazamientos de los nodos
x_truss_bridge_disp = analysis(x_truss_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
x_truss_bridge_esfuerzos = calc_esfuerzos_int(
    x_truss_bridge_data_optimized["nodes"], 
    x_truss_bridge_data_optimized["elements"], 
    x_truss_bridge_data_optimized["mats"], 
    x_truss_bridge_disp
)

"""""
 Optimizar y analizar el puente tipo K
"""""

# Optimizar las secciones del puente Warren para minimizar la masa total
k_truss_bridge_data_optimized, k_truss_bridge_masa_minima = optimizar_cercha(k_truss_bridge_data,
                                                                           max_masa=10, 
                                                                           max_esfuerzo=330, 
                                                                           min_seccion=0.001, 
                                                                           max_seccion=0.02)                                                                             

# Realizar el análisis estructural para obtener los desplazamientos de los nodos
k_truss_bridge_disp = analysis(k_truss_bridge_data_optimized, verbose=False)

# Calcular las fuerzas internas (esfuerzos) en los elementos del puente
k_truss_bridge_esfuerzos = calc_esfuerzos_int(
    k_truss_bridge_data_optimized["nodes"], 
    k_truss_bridge_data_optimized["elements"], 
    k_truss_bridge_data_optimized["mats"], 
    k_truss_bridge_disp
)
```

A continuación se grafican los resultados obtenidos para cada tipo de puente optimizado. Las gráficas muestran los desplazamientos de los nodos y los esfuerzos internos en los elementos después de la optimización. Estas visualizaciones permiten evaluar el comportamiento estructural de cada puente bajo las condiciones de carga especificadas y verificar que las restricciones de diseño se cumplan.

Para cada tipo de puente, se generarán dos gráficos principales:
1. **Desplazamientos de los nodos**: Este gráfico muestra cómo se desplazan los nodos del puente bajo la carga aplicada. Es útil para identificar áreas de mayor deformación y verificar que los desplazamientos estén dentro de los límites aceptables.
2. **Esfuerzos internos en los elementos**: Este gráfico muestra los esfuerzos internos en cada elemento del puente. Es fundamental para asegurar que los esfuerzos no excedan los límites de diseño y que la estructura sea segura.

A continuación se presenta el código utilizado para generar estas visualizaciones.

```python
fig, axs = plt.subplots(3, 2, figsize=(8, 5))

# Visualizar los esfuerzos en la cercha del puente Warren
axs[0, 0].set_title("Warren Bridge - Masa: {:.2f} toneladas".format(warren_bridge_masa_minima), fontsize=8)
plt.sca(axs[0, 0])
vis_esfuerzos_secciones(
    warren_bridge_data_optimized["nodes"], 
    warren_bridge_data_optimized["elements"], 
    warren_bridge_esfuerzos, 
    warren_bridge_data_optimized["mats"], 
    desp=warren_bridge_disp*3
)

# Visualizar los esfuerzos en la cercha del puente Pratt
axs[0, 1].set_title("Pratt Bridge - Masa: {:.2f} toneladas".format(pratt_bridge_masa_minima), fontsize=8)
plt.sca(axs[0, 1])
vis_esfuerzos_secciones(
    pratt_bridge_data_optimized["nodes"], 
    pratt_bridge_data_optimized["elements"], 
    pratt_bridge_esfuerzos, 
    pratt_bridge_data_optimized["mats"], 
    desp=pratt_bridge_disp*3
)

# Visualizar los esfuerzos en la cercha del puente Howe
axs[1, 0].set_title("Howe Bridge - Masa: {:.2f} toneladas".format(howe_bridge_masa_minima), fontsize=8)
plt.sca(axs[1, 0])
vis_esfuerzos_secciones(
    howe_bridge_data_optimized["nodes"], 
    howe_bridge_data_optimized["elements"], 
    howe_bridge_esfuerzos, 
    howe_bridge_data_optimized["mats"], 
    desp=howe_bridge_disp*3
)

axs[1, 1].set_title("Double Warren Bridge - Masa: {:.2f} toneladas".format(double_warren_bridge_masa_minima), fontsize=8)
plt.sca(axs[1, 1])
vis_esfuerzos_secciones(
    double_warren_bridge_data_optimized["nodes"], 
    double_warren_bridge_data_optimized["elements"], 
    double_warren_bridge_esfuerzos, 
    double_warren_bridge_data_optimized["mats"], 
    desp=double_warren_bridge_disp*3
)

# Visualizar los esfuerzos en la cercha del puente K
axs[2, 0].set_title("X Bridge - Masa: {:.2f} toneladas".format(x_truss_bridge_masa_minima), fontsize=8)
plt.sca(axs[2, 0])
vis_esfuerzos_secciones(
    x_truss_bridge_data_optimized["nodes"], 
    x_truss_bridge_data_optimized["elements"], 
    x_truss_bridge_esfuerzos, 
    x_truss_bridge_data_optimized["mats"], 
    desp=x_truss_bridge_disp*3
)

# Visualizar los esfuerzos en la cercha del puente K
axs[2, 1].set_title("K Bridge - Masa: {:.2f} toneladas".format(k_truss_bridge_masa_minima), fontsize=8)
plt.sca(axs[2, 1])
vis_esfuerzos_secciones(
    k_truss_bridge_data_optimized["nodes"], 
    k_truss_bridge_data_optimized["elements"], 
    k_truss_bridge_esfuerzos, 
    k_truss_bridge_data_optimized["mats"], 
    desp=k_truss_bridge_disp*3
)

plt.tight_layout()
plt.savefig("figs/optimized_bridges.png", dpi=300)
plt.show()
```

```python
# Crear una figura con subplots para cada puente
fig, axs = plt.subplots(3, 2, figsize=(10, 5))

# Lista de datos de esfuerzos y títulos para cada puente
puentes = [
    (warren_bridge_esfuerzos, "Warren"),
    (pratt_bridge_esfuerzos, "Pratt"),
    (howe_bridge_esfuerzos, "Howe"),
    (double_warren_bridge_esfuerzos, "Warren Doble"),
    (x_truss_bridge_esfuerzos, "X Truss"),
    (k_truss_bridge_esfuerzos, "K Truss")
]

# Límite de esfuerzo en MPa
limite_esfuerzo = 330

# Iterar sobre cada puente y su respectivo subplot
for i, (esfuerzos, titulo) in enumerate(puentes):
    ax = axs[i // 2, i % 2]
    
    # Colores para los esfuerzos: rojo para positivos y azul para negativos
    colors = ['red' if esfuerzo > 0 else 'blue' for esfuerzo in esfuerzos]
    
    # Crear el gráfico de barras
    ax.bar(range(len(esfuerzos)), esfuerzos/1e6, color=colors)
    
    # Añadir una línea horizontal para el límite de esfuerzo
    ax.axhline(y=limite_esfuerzo, color='gray', linestyle='--', linewidth=1, label=f'Límite de {limite_esfuerzo} MPa')
    
    # Añadir etiquetas y título
    ax.set_xlabel('Elementos', fontsize=8)
    ax.set_ylabel('Esfuerzos (MPa)', fontsize=8)
    ax.set_title(titulo, fontsize=8)
    
    # Añadir leyenda
    #ax.legend(fontsize=8)

# Ajustar el layout para que no se solapen los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
